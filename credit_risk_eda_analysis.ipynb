```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df_dataset.copy()
```

```python
# Renaming the columns for better usability
```

```python
df.columns = ["customer_id" ,"credit_limit","gender","education","marital_status","age","repayment_sep","repayment_aug","repayment_july", 
              "repayment_june", "repayment_may", "repayment_apr","bill_sep","bill_aug", "bill_july","bill_june","bill_may","bill_apr",
              "payment_sep","payment_aug","payment_july","payment_june","payment_may","payment_apr", "default_next_month"]
```

```python
df["gender"] = df["gender"].map({ 1 : "Male" , 2 : "Female"}).fillna("Unknown")
```

```python
print(df["education"].value_counts())
```

```python
df["education"] = df["education"].replace([0,5,6] , np.nan )
```

```python
education_mapping = { 1 : "Graduate" , 2 : "university" , 3 : "High School" , 4 : "Others"  }
```

```python
df["education"] = df["education"].map(education_mapping)
```

```python
df["education"] = df["education"].fillna("Unknown")
```

```python
print(df["education"].value_counts())
```

```python
df["marital_status"] = df["marital_status"].map({ 1 : "Married" , 2 : "Single" , 3 : "Other"}).fillna("Other")
```

```python
df[["gender","education","marital_status"]].head(10)
```

```python
df.head()
```

```python
# Default split across the Education levels
```

```python
edu_default= (
    pd.crosstab( df["education"], df["default_next_month"], normalize="index") * 100 ).round(2)

edu_default.columns = ["No Default (%)", "Default (%)"]
edu_default = edu_default.sort_values(by="Default (%)", ascending=False)
edu_default
```

```python
df["credit_limit_segment"] = pd.qcut(df["credit_limit"], q=4, labels=["Low limit" , "Mid-low", "Mid-high","High limit"])
```

```python
credit_limit_default = ( pd.crosstab(df["credit_limit_segment"],  df["default_next_month"], normalize="index") * 100 ).round(2)

credit_limit_default.columns = ["No Default (%)", "Default (%)"]
credit_limit_default = credit_limit_default.sort_values(by="Default (%)", ascending=False)

credit_limit_default
```

```python
#ploting the graph 
plt.figure(figsize=(8,5))

credit_limit_default["Default (%)"].plot(
    kind="bar",
    color="indigo"
)

plt.title("Default Rate by Credit Limit Segment")
plt.ylabel("Default Percentage")
plt.xlabel("Credit Limit Segment")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Creating age bins 
df["age_group"] = pd.cut(df["age"],bins = [20,30,40,50,60,80], labels=["21-30","31-40","41-50","51-60","60+"], right = False)
df["age_group"]
```

```python
df["age_group"].value_counts().sort_index()
```

```python
# Calculate Default rate by age_group

age_default = (pd.crosstab(df["age_group"], df["default_next_month"],normalize = "index")*100).round(2)
age_default.columns= ["No Default (%)", "Default (%)"]
age_default = age_default.sort_values( by = "Default (%)" , ascending = False)
age_default
```

```python
plt.figure(figsize = (8,5))
age_default["Default (%)"].plot(kind= "bar", color = "indigo")
plt.title("Default rate by age group")
plt.ylabel("Default percentage")
plt.xlabel("Age group")
plt.xticks(rotation = 0)
plt.tight_layout()
plt.savefig("Default_rate_by_age_group.png", dpi=300, bbox_inches= "tight")

plt.show()
```

```python
repayment_months = ["repayment_sep", "repayment_aug","repayment_july","repayment_june","repayment_may","repayment_apr"]
```

```python
df["avg_delay"] = df[repayment_months].mean(axis = 1) 
df["avg_delay"]
```

```python
df["max_delay"] = df[repayment_months].max(axis=1)
df["max_delay"]
```

```python
df["num_delayed_months"] = ((df[repayment_months] > 0 ).sum(axis = 1))
df["num_delayed_months"]
```

```python
# Creating risk segment based on customer behaviour

def risk_segment(value) :
    if value["num_delayed_months"] == 0:
        return "low risk"
    elif value["num_delayed_months"] <= 2:
        return "medium risk"
    else :
        return "high risk"

df["behaviour_segment"] = df.apply(risk_segment, axis = 1)
```

```python
print(df["behaviour_segment"].value_counts())
```

```python
# Default rate by behaviour segment 

behaviour_default = (pd.crosstab( df["behaviour_segment"], df["default_next_month"] , normalize = "index")*100).round(2)
behaviour_default.columns = ["No Default (%)" , "Default (%)"]
behaviour_default = behaviour_default.sort_values(by = "Default (%)" , ascending = False)
behaviour_default
```

```python
plt.figure(figsize = (8,5))
behaviour_default["Default (%)"].plot(kind="bar",color = "indigo")
plt.title("Default rate by behaviour")
plt.ylabel("Default percentage")
plt.xlabel("behaviour segment")
plt.xticks(rotation = 0)
plt.tight_layout()
plt.savefig("Default_rate_by_behaviour.png", dpi=300, bbox_inches= "tight")

plt.show()
```

```python
exposure_summary = df.groupby("credit_limit_segment")["credit_limit"].sum().reset_index()
exposure_summary.columns = ["Credit Limit Segment" , "Total Credit Exposure"]
exposure_summary = exposure_summary.sort_values(by = "Total Credit Exposure" , ascending = False)
exposure_summary
```

```python
df["default_exposure"] = df["credit_limit"] * df["default_next_month"]
exposure_risk = df.groupby("credit_limit_segment")[["credit_limit","default_exposure"]].sum().reset_index()
exposure_risk.columns=["Credit Limit Segment" , "Total Exposure" , "Default Exposure"]
```

```python
# % contribution to total default exposure
exposure_risk["Default Contribution (%)"] = ( (exposure_risk["Default Exposure"] / exposure_risk["Default Exposure"].sum())*100)
exposure_risk = exposure_risk.sort_values(by= "Default Exposure" , ascending = False)
```

```python
plt.figure(figsize = (8,5))
plt.bar(exposure_risk["Credit Limit Segment"] , exposure_risk["Default Contribution (%)"] , color = "indigo")
plt.title("Default Contribution by credit limit segment")
plt.xlabel("Credit Limit Segment")
plt.ylabel("Contribution to total Default (%)")
plt.xticks(rotation = 0)
plt.tight_layout()
plt.savefig("Default_contribution_by_credit_limit_segment.png", dpi=300, bbox_inches= "tight")
plt.show()
```

```python
# Credit Utilization risk
monthly_bills = ["bill_sep", "bill_aug","bill_july","bill_june","bill_may","bill_apr"]
df["avg_bill"] = df[monthly_bills].mean(axis=1)
```

```python
df["utilization_ratio"] = df["avg_bill"]/df["credit_limit"]
```

```python
# Creating Utilization segments 
df["utilization_segments"] = pd.qcut(df["utilization_ratio"] , q=4 , labels=["Low Utili","Mid-Low", "Mid-High", "High Utili"] )
```

```python
utilization_default = ( (pd.crosstab(df["utilization_segments"], df["default_next_month"], normalize = "index")*100)).round(2)
utilization_default.columns = ["No Default (%)", "Default (%)"]  
utilization_default = utilization_default.sort_values(by = "Default (%)" , ascending = False )
utilization_default
```

```python
plt.figure(figsize = (8,5))
utilization_default["Default (%)"].plot( kind= "bar" , color = "indigo")
plt.title("Default Rate by Utilization Segment ")
plt.xlabel("Default (%)")
plt.ylabel("Utilization Segment")
plt.xticks(rotation = 0)
plt.tight_layout()
plt.savefig("Default_rate_by_utilization_segment.png", dpi=300, bbox_inches= "tight")
plt.show()
```

```python
# Track overall payment repayment delay
repayment_months = ["repayment_sep","repayment_aug","repayment_july","repayment_june","repayment_may","repayment_apr"]
```

```python
df["avg_repayment"] = df[repayment_months].mean(axis = 1)
```

```python
df["max_repayment"] = df[repayment_months].max(axis = 1)
```

```python
df["delayed_months"] = (df[repayment_months] > 0).sum(axis = 1)
```

```python
delay_default = (pd.crosstab(df["delayed_months"] ,df["default_next_month"] ,normalize = "index")*100).round(2)
delay_default.columns = ["No Default (%)", "Default (%)"]
delay_default = delay_default.sort_values(by= "Default (%)" , ascending = False)
```

```python
plt.figure(figsize = (8,5))
delay_default["Default (%)"].plot(kind = "bar" , color = "indigo")
plt.title("Default Rate by Number of Delayed Months")
plt.xlabel("Number of Delayed Months" )
plt.ylabel("Default Percentage")
plt.xticks(rotation = 0)
plt.tight_layout()
plt.savefig("Default_delay_by_delayed_months.png", dpi=300, bbox_inches= "tight")
plt.show()
```

```python
# Calculating overall Default Rate 
overall_default_rate = (df["default_next_month"].mean())*100
print("overall_default_rate :",round(overall_default_rate,2) , '%')
```
